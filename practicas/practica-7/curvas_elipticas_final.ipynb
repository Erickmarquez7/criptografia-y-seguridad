# Creación de la clase Punto y Curva eliptica

```python
class Punto:
  def __init__(self, x, y):
    '''Crea un punto en el plano cartesiano'''
    self.x = x
    self.y = y

  def toString(self):
    '''Devuelve la representación en cadena del punto'''
    return f'({self.x}, {self.y})'

  def __str__(self):
    return f'({self.x}, {self.y})'

  def equals(self, otro_punto):
    '''Devuelve True si el punto es igual al otro punto'''
    return self.x == otro_punto.x and self.y == otro_punto.y

  def set(self, nuevo_x, nuevo_y):
    '''Actualiza las coordenadas del punto'''
    self.x = nuevo_x
    self.y = nuevo_y

# Ejemplo de uso
punto1 = Punto(3, 4)
print("Punto 1:", punto1)
```

```python
class CurvaEliptica:
  def __init__(self, a, b, p):
    '''parametros a y b, con p primo'''
    self.a = a
    self.b = b
    self.p = p

  def __str__(self):
    return f'Curva Eliptica(a={self.a}, b={self.b}, p={self.p})'

  def toString(self):
    return f'Curva Eliptica(a={self.a}, b={self.b}, p={self.p})'

  def tiene(self, punto):
    '''nos dice si el punto pertenece a la surva'''
    x, y = punto.x, punto.y
    #basicamente si es solucion a la ecuacion
    return (y**2) % self.p == (x**3 + self.a * x + self.b) % self.p

  def raiz_cuadrada_modular(self, a, p):
    '''raiz cuadrada para obtener todos los puntos'''
    for x in range(p):
      if (x**2) % p == a:
        return x
    return None

  def puntos(self):
    '''obtiene todos los puntos de la curva'''
    puntos = []
    for x in range(self.p):
      y_cuadrado = (x**3 + self.a * x + self.b) % self.p
      raiz_cuadrada_y = self.raiz_cuadrada_modular(y_cuadrado, self.p)
      if raiz_cuadrada_y is not None:
        punto1 = Punto(x, raiz_cuadrada_y)
        punto2 = Punto(x, self.p - raiz_cuadrada_y)
        puntos.extend([punto1, punto2])  # Se agregan objetos Punto
    return puntos

  def orden(self, P):
    if P == 'O':
      return 1  # El orden del punto en el infinito es 1
    for k in range(1, self.p + 1):
      resultado = self.multiplicacion_punto(k, P)
      if resultado == 'O':
        return k
    return None  # Punto no tiene orden conocido dentro del rango p

  def cofactor(self, P):
    orden_p = self.orden_2(P)
    #print(orden_p)
    if orden_p is None or orden_p == 1 or P=='O':
      return 1  # No se pudo determinar el orden o el orden es 1, la curva es buena
    total_puntos = len(self.puntos())+1 #para tomar en cuenta el infinito
    #print('total puntos',total_puntos)
    #print('div', total_puntos/orden_p)
    return total_puntos // orden_p

  def inverso_modular(self, a):
    # Función para calcular el inverso modular de a en F_p
    return pow(a, -1, self.p)

  def suma_puntos(self, P, Q):
    if P == "O":  # Punto en el infinito
      return Q
    if Q == "O":
      return P

    x1, y1 = P.x, P.y
    x2, y2 = Q.x, Q.y

    if x1 == x2 and y1 != y2:
      return "O"  # Punto en el infinito
    elif P != Q:
      m = ((y2 - y1) * self.inverso_modular(x2 - x1)) % self.p
    elif P == Q:
      m = ((3 * x1**2 + self.a) * self.inverso_modular(2 * y1)) % self.p

    x3 = (m**2 - x1 - x2) % self.p
    y3 = (m * (x1 - x3) - y1) % self.p

    return Punto(x3, y3)

  def resta_puntos(self, p, q):
    '''La resta funciona con el inverso P+(-q)'''
    inverso_q = self.punto_inverso(q)
    res=self.suma_puntos(p, inverso_q)
    return res

  def multiplicacion_punto(self, k, P):
    resultado = 'O'  # Punto en el infinito
    for _ in range(k):
        resultado = self.suma_puntos(resultado, P)
    return resultado

  def punto_inverso(self, P):
    if P == 'O':
      return 'O'  # El punto inverso del infinito es el infinito
    x, y = P.x, P.y
    punto_inverso = Punto(x, -y % self.p)
    return punto_inverso

  def teorema_hassee(self):
    # Teorema de Hasse: |E(F_p)| está en el rango [p + 1 - 2 * sqrt(p), p + 1 + 2 * sqrt(p)]
    rango_inferior = self.p + 1 - 2 * int(self.p**0.5)
    rango_superior = self.p + 1 + 2 * int(self.p**0.5)
    return rango_inferior, rango_superior

  def orden_2(self, P):
    if P == 'O':
      return 1  # El orden del punto en el infinito es 1

    rango_inferior, rango_superior = self.teorema_hassee()

    for k in range(rango_inferior, rango_superior + 1):
      resultado = self.multiplicacion_punto(k, P)
      if resultado == 'O':
          return k

    return None  # Punto no tiene orden conocido dentro del rango p

# Ejemplo de uso
#punto_P = Punto(1, 5)
#punto_Q = Punto(5,16)
#curva = CurvaEliptica(2, 2, 17)
#inv = curva.punto_inverso(punto_Q)

curva2 = CurvaEliptica(-2, 8, 233)
punto_G=Punto(0,63)
curva2.orden_2(punto_G)

curva_ejemplo = CurvaEliptica(2, 2, 17)
punto_generador_ejemplo = Punto(5, 1)
#puntos = curva_ejemplo.puntos()
#for i in puntos:
#  print(i)

print('comprobacion xd')
i =1
while i < 20:
  print(curva_ejemplo.multiplicacion_punto(i,punto_generador_ejemplo))
  i=i+1

print('orden:', curva_ejemplo.orden_2(Punto(6,3)))
print('cofactor:', curva_ejemplo.cofactor(Punto(5,1)))
```

# Diffie-Hellman y clase Entidad

```python
import random

class Entidad:
  def __init__(self, nombre, mensaje, curva, punto_generador, tabla_codificacion, tabla_inv, llave_secreta, punto_secreto):
    self.nombre = nombre
    self.mensaje = mensaje
    self.curva = curva
    self.punto_generador = punto_generador
    self.orden_generador = curva.orden_2(punto_generador) # k minimo tal que kP='O'

    self.llave_secreta = llave_secreta #1#random.randint(1, self.orden_generador - 1) #k
    self.punto_secreto = punto_secreto #curva.multiplicacion_punto(self.llave_privada, punto_generador)#k veces el punto generador

    self.llaves_publicas_propias = {}#mis dos llaves publicas
    self.llaves_publicas_otra_entidad = {} #las llaves de la otra entidad
    self.tabla_codificacion = tabla_codificacion #la tabla xd
    #diccionario inverso para codificar el punto a letra, aunque lo mantiene como tupla en lugar de Punto xd
    self.tabla_inv = tabla_inv

  def __str__(self):
    return f"Entidad: {self.nombre}\n" \
      f"Mensaje: {self.mensaje}\n" \
      f"Curva: {self.curva}\n:" \
      f"Punto generador {self.punto_generador}\n " \
      f"Llave Privada: {self.llave_secreta}\n" \
      f"Punto Privado: {self.punto_secreto}\n" \
      f"Llaves Públicas Propias: {self.llaves_publicas_propias}\n" \
      f"Llaves Públicas Otra Entidad: {self.llaves_publicas_otra_entidad}\n" \
      f"Tabla de Codificación: {self.tabla_codificacion}\n"

  def cifrar(self, mensaje):
    '''Manda a llamar el metodo cifrar(numero, caracter) para hacer las parejas de punts cifrados'''
    parejas_puntos_cifrados = []
    for c in mensaje:
      numero = random.randint(0, 24)
      tupla_cifrada= self.cifrar_aux(numero, c)
      parejas_puntos_cifrados.append(tupla_cifrada)
    return parejas_puntos_cifrados

  def cifrar_aux(self, numero, caracter):
    '''
    este metodo es para comprobar que el metodo cifra correctamente dado un numero
    ya que en la implementacion original es con numero aleatorio y seria complicado comprobar a mano
    '''
    #numero es el numero aleatorio
    e1=self.curva.multiplicacion_punto(numero, self.punto_generador)
    M=self.tabla_codificacion[caracter]
    # t de termino xd
    t2=self.curva.multiplicacion_punto(self.llave_secreta+numero, self.llaves_publicas_otra_entidad['pk1']) #(b+r)A1
    t3=self.curva.multiplicacion_punto(numero, self.llaves_publicas_otra_entidad['pk2'])#r(A2)
    t4=self.curva.resta_puntos(t2, t3)#(b+r)A1 - r(A2)

    t5=self.curva.suma_puntos(M, t4)#M+(b+r)A1 - r(A2)
    e2=self.curva.suma_puntos(t5, self.llaves_publicas_otra_entidad['pke'])
    puntos = e1, e2
    t1=e1.x, e1.y
    t2=e2.x, e2.y
    return self.tabla_inv[t1], self.tabla_inv[t2]

  def descifrar(self, parejas_puntos):
    '''Decifra el mensaje utiliando la tabla inversa y el metodo descifrar Aux'''
    mensaje=''
    for t in parejas_puntos:
      letra = self.descifrar_aux(t)
      mensaje=mensaje+letra
    return mensaje


  def descifrar_aux(self, tupla):
    e1 = self.tabla_codificacion[tupla[0]]
    e2 = self.tabla_codificacion[tupla[1]]
    t1 = self.curva.multiplicacion_punto(self.llave_secreta, e1)
    t2 = self.curva.multiplicacion_punto(self.llave_secreta, self.llaves_publicas_otra_entidad['pk1'])
    t3 = self.curva.suma_puntos(t1,t2)
    t4 = self.curva.suma_puntos(t3, self.llaves_publicas_otra_entidad['pke'])
    M = self.curva.resta_puntos(e2, t4)
    t=M.x, M.y
    return self.tabla_inv[t]

  def genera_llaves_publicas(self):
    #self.alpha = 1#random.randint(1,self.orden_generador-1)
    suma=self.curva.suma_puntos(self.punto_generador, self.punto_secreto)
    pk1 = self.curva.multiplicacion_punto(self.llave_secreta, suma) #A1 = a(C+A)
    pk2 = self.curva.multiplicacion_punto(self.llave_secreta, self.punto_secreto) #A2=a(A), donde C es el generador y A el privado
    self.llaves_publicas_propias['pk1']=pk1
    self.llaves_publicas_propias['pk2']=pk2

  def recibe_llaves_publicas(self, pks):
    #print('MIS LLAVES en recibir',self.llaves_publicas_propias['pk1'])
    self.llaves_publicas_otra_entidad = pks #las llaves de la otra entidad
    llave_final = self.curva.multiplicacion_punto(self.llave_secreta, pks['pk2']) #alpha por B2
    self.llaves_publicas_propias['pke']=llave_final #agregamos la tercer llave publica
    return llave_final #se la pasamos a bob
    #return {'pk1': self.llaves_publicas_propias['pk1'], 'pk2': llave_final, 'pke': self.punto_privado}

  def recibe_llave_final(self, kf):
    self.llaves_publicas_otra_entidad['pke']=kf

curva_eliptica = CurvaEliptica(2,9,37)
punto_generador = Punto(9,4)

punto_privado_alice = Punto(10,20)
punto_privado_bob = Punto(11,20)

k_a=5
k_b=7

tabla = {
  'a': Punto(5, 25),
  'b': Punto(1, 30),
  'c': Punto(21, 32),
  'd': Punto(7, 25),
  'e': Punto(25, 12),
  'f': Punto(4, 28),
  'g': Punto(0, 34),
  'h': Punto(16, 17),
  'i': Punto(15, 26),
  'j': Punto(27, 32),
  'k': Punto(9, 4),
  'l': Punto(2, 24),
  'm': Punto(26, 5),
  'n': Punto(33, 14),
  'o': Punto(11, 17),
  'p': Punto(31, 22),
  'q': Punto(13, 30),
  'r': Punto(35, 21),
  's': Punto(23, 7),
  't': Punto(10, 17),
  'u': Punto(29, 6),
  'v': Punto(29, 31),
  'w': Punto(10, 20),
  'x': Punto(23, 30),
  'y': Punto(35, 16),
  'z': Punto(13, 7),
  '1': Punto(31, 15),
  '2': Punto(11, 20),
  '3': Punto(33, 23),
  '4': Punto(26, 32),
  '5': Punto(2, 13),
  '6': Punto(9, 33),
  '7': Punto(27, 5),
  '8': Punto(15, 11),
  '9': Punto(16, 20),
  '0': Punto(0, 3),
  '#': Punto(4, 9),
  '@': Punto(25, 25),
  '!': Punto(7, 12),
  '&': Punto(21, 5),
  '$': Punto(1, 7),
  '%': Punto(5, 12),
  '*': 'O',
}

tabla_inv = {
  (5, 25) : 'a',
  (1, 30) : 'b',
  (21, 32) : 'c',
  (7, 25) : 'd',
  (25, 12) : 'e',
  (4, 28) : 'f',
  (0, 34) : 'g',
  (16, 17) : 'h',
  (15, 26) : 'i',
  (27, 32) : 'j',
  (9, 4)  : 'k',
  (2, 24) : 'l',
  (26, 5) : 'm',
  (33, 14) : 'n',
  (11, 17) : 'o',
  (31, 22) : 'p',
  (13, 30) : 'q',
  (35, 21) : 'r',
  (23, 7) : 's',
  (10, 17) : 't',
  (29, 6) : 'u',
  (29, 31) : 'v',
  (10, 20) : 'w',
  (23, 30) : 'x',
  (35, 16) : 'y',
  (13, 7) : 'z',
  (31, 15) : '1',
  (11, 20) : '2',
  (33, 23) : '3',
  (26, 32) : '4',
  (2, 13) : '5',
  (9, 33) : '6',
  (27, 5) : '7',
  (15, 11) : '8',
  (16, 20) : '9',
  (0, 3)  : '0',
  (4, 9)  : '#',
  (25, 25) : '@',
  (7, 12) : '!' ,
  (21, 5) : '&' ,
  (1, 7)  : '$' ,
  (5, 12) : '%' ,
  'O' :  '*'
}

#def __init__(self, nombre, mensaje, curva, punto_generador, tabla_codificacion):

alice=Entidad('alice', 'msj_a', curva_eliptica, punto_generador, tabla, tabla_inv, k_a, punto_privado_alice)

bob=Entidad('bob', 'msj_b', curva_eliptica, punto_generador, tabla, tabla_inv, k_b, punto_privado_bob)
```
